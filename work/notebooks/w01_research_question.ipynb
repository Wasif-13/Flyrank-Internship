## 2. The question: decision, action, cost of a wrong call

## 2. The question: decision, action, cost of a wrong call

### Research Question

Which content pages should be reviewed first for refresh based on their search visibility, engagement, and recent performance?

### Decision

Determine whether a content page should be prioritized for review.

### Unit of Analysis

One row represents one content page.

### Output

A ranked list of pages ordered by their refresh priority.

### Action

The content team reviews the highest-ranked pages first and decides whether to refresh, rewrite, expand, monitor, or leave the page unchanged.

### Cost of a Wrong Recommendation

If an important page is missed, valuable traffic opportunities may be lost. If an unnecessary page is recommended, reviewers spend time on pages that may not produce meaningful improvements.

### Why Data or Machine Learning Can Help

The dataset contains observable signals such as impressions, clicks, CTR, sessions, freshness, and content age. These measurements may reveal patterns that help prioritize pages for review. Machine learning supports this decision but does not replace human judgment.
